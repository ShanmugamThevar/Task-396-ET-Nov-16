```python
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException, WebDriverException, TimeoutException
import pandas as pd
```

```python
driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://economictimes.indiatimes.com/archive/year-2016,month-11.cms')
```

```python
urls=[]
date=driver.find_elements_by_xpath("//span[@class='normtxt']/div/table/tbody/tr/td/a[1]")
for i in date:
    urls.append(i.get_attribute('href'))
    
print(len(urls))
```

```python
nurl=urls[2:]
print(len(nurl))
```

```python
news=[]
for i in nurl:
    driver.get(i)
    try:
        news1=driver.find_elements_by_xpath("//ul[@class='content']/li/a")
        for i in news1:
            news.append(i.get_attribute('href'))
    except TimeoutException:
        news.append('NA')
    except NoSuchElementException:
        news.append('NA')
        
print(len(news))
```

```python
news1=news[0:1500]
```

```python
date=[]
author=[]
vertical=[]
headline=[]
describe=[]
for i in news1:
    driver.get(i)
    try:
        lines=driver.find_element_by_xpath("//h1[@class='artTitle font_faus']")
        headline.append(lines.text)
    except WebDriverException:
        headline.append('NA')
    except NoSuchElementException:
        headline.append('NA')
    try:
        authors=driver.find_element_by_xpath("//span[@class='ag']")
        author.append(authors.text)
    except WebDriverException:
        author.append('ET')
    except NoSuchElementException:
        author.append('ET')
    try:
        dates=driver.find_element_by_xpath("//div[@class='dtc vam artByline']/time")
        date.append(dates.text)
    except WebDriverException:
        date.append('NA')
    except NoSuchElementException:
        date.append('NA')
    try:
        vert=driver.find_element_by_xpath("//div[@class='clr breadCrumb contentwrapper']/span[3]")
        vertical.append(vert.text)
    except WebDriverException:
        vertical.append('Other Image')
    except NoSuchElementException:
        vertical.append('Other Image')
    try:
        descr=driver.find_element_by_xpath("//h2[@class='summary']")
        describe.append(descr.text)
    except WebDriverException:
        describe.append('NA')
    except NoSuchElementException:
        describe.append('NA')
        
        
        
        
print(len(headline),len(author),len(date),len(vertical),len(describe))
```

```python
News=pd.DataFrame({'Date & Time':date,'Headlines':headline,'Author':author,'Vertical':vertical,'Description':describe})
News
```

```python
News.to_csv('News1.csv')
```

```python
News.to_json('News1.json')
```
